### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





### Data Description

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


## Limit Data to Bar Coupons

```python
#Limit data to Bar Coupons and Preview
bar_data = data.query('coupon == "Bar"')
bar_data.head(5)
```

## Explore Bar Coupon Data

2. What proportion of bar coupons were accepted?


```python
#Determine the percentage of Bar copuons that were accepted
bar_coupon_accepted_ratio = bar_data.query('Y==1')['Y'].shape[0] / bar_data.shape[0]
print ("Overall proportion that accepted the Bar coupon was {:.2%}".format(bar_coupon_accepted_ratio))
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
bar_data_accepted_lte3 = bar_data.query("Bar in ('never', 'less1', '1~3')").groupby('Y')['Y'].count()
bar_accepted_lte3_ratio = bar_data_accepted_lte3[1]/(bar_data_accepted_lte3[0]+bar_data_accepted_lte3[1])
print("Proportion that went to the bar <= 3 times per month and accepted the Bar coupon was {:.2%}".format(bar_accepted_lte3_ratio))

bar_data_accepted_gt3 = bar_data.query("Bar in ('gt8','4~8')").groupby('Y')['Y'].count()
bar_accepted_gt3_ratio = bar_data_accepted_gt3[1]/(bar_data_accepted_gt3[0]+bar_data_accepted_gt3[1])
print("Proportion that went to the bar  > 4 times per month and accepted the Bar coupon was {:.2%}".format(bar_accepted_gt3_ratio))

```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
bar_visit_gt1_agegt25 = bar_data.query("Y==1" and "Age > 25" and "Bar in ('1~3','gt8','4~8')").groupby('Y')['Y'].count()
bar_visit_gt1_agegt25_ratio = bar_visit_gt1_agegt25[1]/(bar_visit_gt1_agegt25[0]+bar_visit_gt1_agegt25[1])
print("Proportion that went to the bar  > 1X/month & were over 25 who accepted the Bar coupon was {:.2%}".format(bar_visit_gt1_agegt25_ratio ))

bar_visit_lt1_agegt25 = bar_data.query("Y==1" and "Age > 25" and "Bar not in ('1~3','gt8','4~8')").groupby('Y')['Y'].count()
bar_visit_lt1_agegt25_ratio = bar_visit_lt1_agegt25[1]/(bar_visit_lt1_agegt25[0]+bar_visit_lt1_agegt25[1])
print("Proportion that went to the bar  < 1X/month & were over 25 who accepted the Bar coupon was {:.2%}".format(bar_visit_gt1_agegt25_ratio ))
print("Observation: for people who went more than once a month, age was not a factor in the acceptance rate.")
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
bar_data_accepted_gt1mon_wnokid_nowid_occlist = bar_data.query("Bar in ('1~3','gt8','4~8')" and "passenger in ('Friend(s)','Partner')" and "occupation not in ('Farming Fishing & Forestry')" ).groupby('Y')['Y'].count()
bar_data_accepted_gt1mon_wnokid_nowid_occlist_ratio = bar_data_accepted_gt1mon_wnokid_nowid_occlist[1]/(bar_data_accepted_gt1mon_wnokid_nowid_occlist[0]+bar_data_accepted_gt1mon_wnokid_nowid_occlist[1])
print("Proportion that went to the bar more than once/month, had no kid passengers, and did not work in farming, fishing, or forestry who accepted the Bar coupon was {:.2%}".format(bar_data_accepted_gt1mon_wnokid_nowid_occlist_ratio))
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
bar_data_accepted_gt1mon_wnokid_nowid = bar_data.query("Bar in ('1~3','gt8','4~8')" and "passenger in ('Friend(s)','Partner')" and "maritalStatus != 'Widowed'" ).groupby('Y')['Y'].count()
bar_data_accepted_gt1mon_wnokid_nowid_ratio = bar_data_accepted_gt1mon_wnokid_nowid[1]/(bar_data_accepted_gt1mon_wnokid_nowid[0]+bar_data_accepted_gt1mon_wnokid_nowid[1])
print("Proportion that went to the bar more >1/month, had no kid passengers, and were not widowed who accepted the Bar coupon was {:.2%}".format(bar_data_accepted_gt1mon_wnokid_nowid_ratio))

bar_data_accepted_gt1mon_lt30 = bar_data.query("Bar in ('1~3','gt8','4~8')" and "age in ('21','26','below21')").groupby('Y')['Y'].count()
bar_data_accepted_gt1mon_lt30_ratio = bar_data_accepted_gt1mon_lt30[1]/(bar_data_accepted_gt1mon_lt30[0]+bar_data_accepted_gt1mon_lt30[1])
print("Proportion that went to the bar more >1/month and were less than 30 who accepted the Bar coupon was {:.2%}".format(bar_data_accepted_gt1mon_lt30_ratio))

bar_data_accepted_gt4times_ltinc50 = bar_data.query("RestaurantLessThan20 in ('4~8','gt8')" and "income in ('$37500 - $49999','$25000 - $37499')").groupby('Y')['Y'].count()
bar_data_accepted_gt4times_ltinc50_ratio = bar_data_accepted_gt4times_ltinc50[1]/(bar_data_accepted_gt4times_ltinc50[0]+bar_data_accepted_gt4times_ltinc50[1])
print("Proportion that went to cheap restaraunts >4X/month and have an income <$50k who accepted the Bar coupon was {:.2%}".format(bar_data_accepted_gt4times_ltinc50_ratio))
```

### Comparing Destinations

```python
sns.displot(data=coffee_house_data,x='destination',multiple='dodge',hue='Y')
plt.xlabel("Destination")
plt.ylabel("Count of Accepted(1) or Declined(0)")
plt.xticks(rotation=45)
plt.title("Comparing Acceptance by Destination")
plt.show()
```

```python
coffee_house_accepted_destnourgent = coffee_house_data.query("destination == 'No Urgent Place'").groupby('Y')['Y'].count()
coffee_house_accepted_destnourgent_ratio = coffee_house_accepted_destnourgent[1]/(coffee_house_accepted_destnourgent[0]+coffee_house_accepted_destnourgent[1])
print("Proportion of people with no urgent destination that accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_destnourgent_ratio))

coffee_house_accepted_desthome = coffee_house_data.query("destination == 'Home'").groupby('Y')['Y'].count()
coffee_house_accepted_desthome_ratio = coffee_house_accepted_desthome[1]/(coffee_house_accepted_desthome[0]+coffee_house_accepted_desthome[1])
print("Proportion of people with heading home that accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_desthome_ratio))

coffee_house_accepted_destwork = coffee_house_data.query("destination == 'Work'").groupby('Y')['Y'].count()
coffee_house_accepted_destwork_ratio = coffee_house_accepted_destwork[1]/(coffee_house_accepted_destwork[0]+coffee_house_accepted_destwork[1])
print("Proportion of people with heading to work that accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_destwork_ratio))

```

```python
sns.displot(data=coffee_house_data,x='income',multiple='dodge',hue='Y')
plt.xlabel("Income")
plt.xticks(rotation=45)
plt.ylabel("Count of Accepted(1) or Declined(0)")
plt.title("Comparing Acceptance by Income")
plt.show()
```

```python
coffee_house_accepted_inclt75 = coffee_house_data.query("income in ('$37500 - $49999','$62500 - $74999','$12500 - $24999','$50000 - $62499' '$25000 - $37499')").groupby('Y')['Y'].count()
coffee_house_accepted_inclt75_ratio = coffee_house_accepted_inclt75[1]/(coffee_house_accepted_inclt75[0]+coffee_house_accepted_inclt75[1])
print("Proportion of people with income < $75K that accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_inclt75_ratio))

coffee_house_accepted_incgte75 = coffee_house_data.query("income in ('$75000 - $87499','$87500 - $99999','$100000 or More')").groupby('Y')['Y'].count()
coffee_house_accepted_incgte75_ratio = coffee_house_accepted_incgte75[1]/(coffee_house_accepted_incgte75[0]+coffee_house_accepted_incgte75[1])
print("Proportion of people with income > $75K that accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_incgte75_ratio))

```

```python
sns.displot(data=coffee_house_data,x='occupation',multiple='dodge',hue='Y',height=4, aspect=4.5)
plt.xlabel("Occupation")
plt.title("Comparing Acceptance by Occupation")
plt.xticks(rotation=75)
plt.ylabel("Count of Accepted(1) or Declined(0)")
plt.show()

```

```python
coffee_house_accepted_student = coffee_house_data.query("occupation == 'Student'").groupby('Y')['Y'].count()
coffee_house_accepted_student_ratio = coffee_house_accepted_student[1]/(coffee_house_accepted_student[0]+coffee_house_accepted_student[1])
print("Proportion of Students that accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_student_ratio))

coffee_house_accepted_unemployed = coffee_house_data.query("occupation == 'Unemployed'").groupby('Y')['Y'].count()
coffee_house_accepted_unemployed_ratio = coffee_house_accepted_unemployed[1]/(coffee_house_accepted_unemployed[0]+coffee_house_accepted_unemployed[1])
print("Proportion of unemployed people that accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_unemployed_ratio))

coffee_house_accepted_sales = coffee_house_data.query("occupation == 'Sales & Related'").groupby('Y')['Y'].count()
coffee_house_accepted_sales_ratio = coffee_house_accepted_sales[1]/(coffee_house_accepted_sales[0]+coffee_house_accepted_sales[1])
print("Proportion of 'Sales and related' people that accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_sales_ratio))
```

```python
coffee_house_accepted_nourg_lt75_student = coffee_house_data.query("destination == 'No Urgent Place'" and "income in ('$37500 - $49999','$62500 - $74999','$12500 - $24999','$50000 - $62499' '$25000 - $37499')" and "occupation == 'Student'").groupby('Y')['Y'].count()
coffee_house_accepted_nourg_lt75_student_ratio = coffee_house_accepted_nourg_lt75_student[1]/(coffee_house_accepted_nourg_lt75_student[0]+coffee_house_accepted_nourg_lt75_student[1])
print("Proportion of Students that make less than 75K and had no urgent place to be who accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_nourg_lt75_student_ratio))

coffee_house_accepted_home_gt75_sales = coffee_house_data.query("destination == 'Home'" and "income in ('$75000 - $87499','$87500 - $99999','$100000 or More'" and "occupation == 'Sales & Related'").groupby('Y')['Y'].count()
coffee_house_accepted_home_gt75_sales_ratio = coffee_house_accepted_home_gt75_sales[1]/(coffee_house_accepted_home_gt75_sales[0]+coffee_house_accepted_home_gt75_sales[1])
print("Proportion of 'Sales & Related'people that make mores than 75K and headed home who accepted the Coffee House coupon was {:.2%}".format(coffee_house_accepted_home_gt75_sales_ratio))
```

## Coffee House Coupon Acceptance Observations
